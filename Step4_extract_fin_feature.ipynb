# 提取背鳍特征



```python
import os
import glob
import timm
import pandas as pd
import torchvision.transforms as T

from wildlife_tools.features import DeepFeatures
from wildlife_tools.data import ImageDataset
```

```python
root_dir = r'/media/filming/2025-白海豚/20240825-JM_02-1/'
metainfo = pd.read_csv(root_dir + "FIN_METAINFO.csv", index_col=0)
```

```python
model = timm.create_model('hf-hub:BVRA/MegaDescriptor-L-384', pretrained=True)
```

```python
transform = T.Compose([
    T.Resize([384, 384]),
    T.ToTensor(),
    T.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225))
])
dataset = ImageDataset(
        root = root_dir,
        metadata=metainfo.query('select==True'),
        transform=transform,
)
```

```python
extractor = DeepFeatures(model, device='cuda', batch_size=32) 
# batch size 32 to match 12G gpu memory
features = extractor(dataset)
```

```python
features.save(root_dir + "/deepfeatures")
```
